```python
#load necessary module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
normal= np.random.rand(12,15)
sns.heatmap(normal)
```

```python
sns.heatmap(normal, annot= True)
```

```python
sns.heatmap(normal,vmin=0, vmax=2)
```

```python
sns.heatmap(normal,vmin=0, vmax=3)
```

```python
#Load dataset 
flights= sns.load_dataset('flights')
flights.head()
```

```python
flights.tail()
```

```python
#Now we will do pivot our flights dataset to see clearly 
flights= flights.pivot('month','year','passengers')
sns.heatmap(flights)
```

```python
sns.heatmap(flights, annot= True)
```

```python
sns.heatmap(flights, annot= True, fmt='d')
```

```python
sns.heatmap(flights, linewidth=0.9)
```

```python
sns.heatmap(flights, linewidth=0.9, annot= True, fmt='d')
```

```python
#Now see the min and max value 

sns.heatmap(flights, linewidth=0.9, annot= True, fmt='d', vmin=100, vmax=650)
```

```python
sns.heatmap(flights, cmap='RdBu', annot= True, fmt='d')
```

```python
sns.heatmap(flights, cmap='summer', annot= True, fmt='d')
```

```python
sns.heatmap(flights, cmap='winter_r', annot= True, fmt='d')
```

```python
sns.heatmap(flights, cmap='coolwarm', annot= True, fmt='d')
```

```python
#Marks specific dataset 
sns.heatmap(flights, center= flights.loc['June',1954], annot= True, fmt='d')
```

```python
sns.heatmap(flights, center= flights.loc['March',1959], annot= True, fmt='d')
```

```python
sns.heatmap(flights, cbar= False)
```

```python
sns.heatmap(flights, cbar= False, linewidths= 1.1)
```
